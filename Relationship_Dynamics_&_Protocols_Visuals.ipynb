```python
!pip install graphviz
!apt-get install -y graphviz  # Install Graphviz system package

from graphviz import Digraph
from IPython.display import display, Image

def create_decision_tree():
    tree = Digraph(comment='Family Planning and Crisis Response Decision Tree')
    tree.attr(size='20,20', bgcolor='lightgrey', style='filled', color='white', fontname="Helvetica")

    # Starting Point
    tree.node('A', 'Identify Life Event', shape='ellipse', color='black', style='filled', fillcolor='lightblue')

    # Dating Branch
    tree.node('B', 'Dating', shape='box', color='darkgreen')
    tree.edge('A', 'B')
    tree.node('B1', 'Separate Residences:\nBryan: Own Residence & Upkeep\nRachel: Own Residence & Upkeep', shape='box', color='darkgreen')
    tree.edge('B', 'B1')
    tree.node('B3', 'Child-Related Costs:\nBryan: None\nRachel: All Child Expenses', shape='box', color='darkgreen')
    tree.edge('B', 'B3')
    tree.node('B4', "Rachel: Separate Accounts for Children\nChildren: Debit Cards, Access to Funds", shape='box', color='darkgreen')
    tree.edge('B3', 'B4')
    tree.node('B5', "Extenuating Circumstances:\nBryan's Help Appreciated, Not Expected\nRachel's Family/Friends Assist", shape='box', color='darkgreen')
    tree.edge('B3', 'B5')

    # Engaged Branch
    tree.node('C', 'Engaged', shape='box', color='darkgreen')
    tree.edge('A', 'C')
    tree.node('C1', 'Cohabiting:\nShared Costs (See Chart)\nRachel: Utilities, Groceries, etc.\nBryan: Major Household Costs', shape='box', color='darkgreen')
    tree.edge('C', 'C1')
    tree.node('C3', 'Child-Related Costs:\nBryan: None\nRachel: All Child Expenses', shape='box', color='darkgreen')
    tree.edge('C', 'C3')
    tree.node('C4', "Rachel: Separate Accounts for Children\nChildren: Debit Cards, Access to Funds", shape='box', color='darkgreen')
    tree.edge('C3', 'C4')
    tree.node('C5', "Extenuating Circumstances:\nBryan's Help Appreciated, Not Expected\nRachel's Family/Friends Assist", shape='box', color='darkgreen')
    tree.edge('C3', 'C5')

    # Married Branch (Similar structure to Engaged)
    tree.node('D', 'Married', shape='box', color='darkgreen')
    tree.edge('A', 'D')
    tree.node('D1', 'Cohabiting:\nShared Costs (See Chart)\nRachel: Utilities, Groceries, etc.\nBryan: Major Household Costs', shape='box', color='darkgreen')
    tree.edge('D', 'D1')
    tree.node('D3', 'Child-Related Costs:\nBryan: None\nRachel: All Child Expenses', shape='box', color='darkgreen')
    tree.edge('D', 'D3')
    tree.node('D4', "Rachel: Separate Accounts for Children\nChildren: Debit Cards, Access to Funds", shape='box', color='darkgreen')
    tree.edge('D3', 'D4')
    tree.node('D5', "Extenuating Circumstances:\nBryan's Help Appreciated, Not Expected\nRachel's Family/Friends Assist", shape='box', color='darkgreen')
    tree.edge('D3', 'D5')

    # Family Planning Branch
    tree.node('FP', 'Family Planning', shape='box', color='darkgreen')
    tree.edge('A', 'FP')
    tree.node('FP1', 'Existing Children', shape='box', color='darkgreen')
    tree.edge('FP', 'FP1')
    tree.node('FP2', 'Preparing to Leave Nest:\n- Timeline\n- Responsibilities', shape='box', color='darkgreen')
    tree.edge('FP1', 'FP2')  # Corrected line

    return tree

decision_tree = create_decision_tree()
png_data = decision_tree.pipe(format='png')
display(Image(png_data))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define relationship stages and years
stages = ['Engagement', 'Marriage', 'Family Planning (Year 1)', 'Family Planning (Year 2)', 'Family Planning (Year 3)']
years = np.arange(len(stages))

# Monthly financial contributions (estimates)
rachel_child_support = [362, 362, 362, 362, 362]  # Rachel's child support (constant)
rachel_travel = [1000, 1000, 1000, 1000, 1000]  # Rachel's travel for child visits (average)
rachel_utilities = [400, 200, 200, 200, 200]  # Rachel's utilities (shared after marriage)
rachel_groceries = [300, 150, 150, 150, 150]  # Rachel's groceries (shared after marriage)
bryan_housing = [2500, 1250, 1250, 1250, 1250]  # Bryan's housing (shared after marriage)
bryan_household = [200, 100, 100, 100, 100]  # Bryan's household expenses (shared after marriage)

# Calculate total contributions
rachel_total = np.array(rachel_child_support) + np.array(rachel_travel) + np.array(rachel_utilities) + np.array(rachel_groceries)
bryan_total = np.array(bryan_housing) + np.array(bryan_household)
combined_total = rachel_total + bryan_total

# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(years, rachel_total, label="Rachel's Total", marker='o', color='lightcoral')
plt.plot(years, bryan_total, label="Bryan's Total", marker='o', color='lightblue')
plt.plot(years, combined_total, label="Combined Total", marker='o', color='darkgreen')

# Customize the graph
plt.xticks(years, stages, rotation=45, ha='right')
plt.ylabel("Monthly Financial Contribution ($)")
plt.title("Scenario Analysis of Financial Contributions Across Relationship Stages")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define relationship stages and years
stages = ['Engagement', 'Marriage', 'Family Planning (Year 1)']
years = np.arange(len(stages))

# --- Directly assign monthly totals ---
rachel_total = np.array([2262, 2262, 1712])  # Rachel's total, including $1412 in Year 1
bryan_total = np.array([2700, 2700, 3450])  # Bryan's total, including $3450 in Year 1

# Calculate the combined total correctly
combined_total = rachel_total + bryan_total

# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(years, rachel_total, label="Rachel's Total", marker='o', color='lightcoral')
plt.plot(years, bryan_total, label="Bryan's Total", marker='o', color='lightblue')
plt.plot(years, combined_total, label="Combined Total", marker='o', color='darkgreen')

# --- Add numbers on the dots ---
for i, txt in enumerate(rachel_total):
    plt.text(years[i], rachel_total[i], f"${txt:.0f}", ha='center', va='bottom')
for i, txt in enumerate(bryan_total):
    plt.text(years[i], bryan_total[i], f"${txt:.0f}", ha='center', va='top')
for i, txt in enumerate(combined_total):
    plt.text(years[i], combined_total[i], f"${txt:.0f}", ha='center', va='center')

# Customize the graph
plt.xticks(years, stages, rotation=45, ha='right')  # Set x-axis labels
plt.ylabel("Monthly Financial Contribution ($)")
plt.title("Scenario Analysis of Financial Contributions Across Relationship Stages")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
